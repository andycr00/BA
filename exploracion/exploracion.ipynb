#### Top 10 de paises por ventas

```python
fact_rental.set_index('sk_location').join(dim_location.set_index('pk_location'))['country'].value_counts().head(10).plot.barh()
```

#### Top 10 de ciudades por ventas

```python
fact_rental.set_index('sk_location').join(dim_location.set_index('pk_location')).groupby(['country', 'city'])['rental_value'].sum().sort_values(ascending=False).head(15).plot.barh()
```

#### Top de clientes por valor vendido

```python
fact_rental.set_index('sk_customer').join(dim_customer.set_index('pk_customer')).groupby(['first_name', 'last_name'])['rental_value'].sum().sort_values(ascending=False).head(20).plot.barh()
plt.ylabel('Clientes')
plt.xlabel('Total Comprado')
```

#### Top de clientes por frecuencia

```python
a = fact_rental.set_index('sk_customer').join(dim_customer.set_index('pk_customer'))['email'].value_counts().head(20).reset_index()
a['value'] = a['email']
a['email'] = a['index']
a = a.drop(columns='index')

b = pd.merge(a, dim_customer, on='email')[['first_name', 'last_name', 'value']]
b['name'] = b['first_name'] + ' ' + b['last_name']
b = b.drop(columns=['first_name', 'last_name'])
b.plot.barh(x='name', y='value')
plt.ylabel('Clientes')
plt.xlabel('Total Comprado')
```

#### Vemos el comportamiento del total de clientes

```python
plt.figure(figsize=(25, 5))
fact_rental.set_index('sk_customer').join(dim_customer.set_index('pk_customer'))['email'].value_counts().rolling(1).mean().plot()
plt.ylabel('Compras realizadas')
plt.xlabel('Clientes')
```

#### Podemos ver un comportamiento similar de las peliculas

#### Top de Categorias de peliculas

```python
plt.figure(figsize=(30, 5))
fact_rental.set_index('sk_film').join(dim_film.set_index('pk_film'))['category'].value_counts().head(20).plot.bar()
```

#### La distribucion estadistica de la tabla rental

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axe = axes.ravel()

fact_rental['rental_value'].describe().drop(['count']).plot(kind='bar', ax=axe[0], title='Rental Value')
dim_film['rental_rate'].describe().drop(['count']).plot(kind='bar', ax=axe[1], title='Rental Rate film')
```

### Histograma de los valores de cada renta y sus calificaciones
